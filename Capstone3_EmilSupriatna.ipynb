## A. Business Problem

**Context**

Dalam persaingan bisnis e-commerce, kasus kehilangan pelanggan (*Customers Churn*) karena beralih ke pesaing atau memutuskan untuk tidak lagi menggunakan layanan e-commerce akan sangat merugikan, terutama dari sisi income perusahaan. Sebab, mendatangkan pelanggan baru melalui strategi marketing membutuhkan biaya yang tidak sedikit, sehingga pelanggan yang sudah ada diharapkan dapat dipertahankan. Di sisi lain, mempertahankan seluruh pelanggan dengan selalu memberikan promosi juga dapat menguras biaya. Dengan demikian, diperlukan strategi khusus untuk mempertahankan pelanggan dengan cara memberikan promosi yang lebih tersegmentasi.

Pada kasus ini, sebuah perusahaan e-commerce meminta tim Data Scienctist untuk membuat prediksi yang dapat memberikan peringatan terhadap pelanggan-pelanggan yang akan *churn* dengan cara menganalisis faktor-faktor penyebab *Churn* berdasarkan histori pelanggan-pelanggan sebelumnya. Sehingga, tim promosi dapat menawarkan promosi yang lebih terorientasi pada jenis pelanggan ini sebelum mereka benar-benar *churn*. Promosi ini diharapkan dapat mendorong pelanggan untuk tetap menggunakan layanan e-commerce perusahaan ini.

**Problem Statement**

Pelanggan merupakan aset berharga bagi suatu bisnis ecommerce. Mempertahankan pelanggan dapat berarti investasi jangka panjang bagi income perusahaan. Oleh karena itu, mempertahankan pelanggan dengan tetap mendorong value transaksi yang meningkat atau setidaknya stabil merupakan hal yang harus dilakukan oleh perusahaan e-commerce melalui tim promosi dan tim sales, di samping terus meningkatkan jumlah pelanggan baru oleh tim marketing, sehingga perusahaan dapat bertumbuh secara progresif.

Sayangnya, pelanggan bersifat fluktuatif (tidak stabil). Pelanggan dapat dengan mudah beralih dari satu layanan ecommerce ke layanan ecommerce lainnya, kecuali mereka sudah menemukan kenyamanan pada satu ecommerce. Hilangnya pelanggan (*Customer Churn*) dapat berarti juga hilangnya potensi income perusahaan, terutama jika value transaksi dari pelanggan tersebut terbilang besar. Pemberian promosi kepada seluruh pelanggan dengan harapan mereka tetap menggunakan layanan ecommerce dapat sangat membebani keuangan perusahaan, namun jika promosi diberikan secara acak juga tidak akan efektif untuk mengatasi *Customer Churn*. Oleh sebab itu, diperlukan prediksi terhadap pelanggan-pelanggan yang cenderung akan *churn*, sehingga promosi dapat diberikan secara efisien dan efektif, terutama dari segi biaya, untuk mengurangi jumlah kasus *Customer Churn* pada perusahaan ecommerce dalam penelitian ini.

**Goals**

Berdasarkan permasalahan tersebut, perusahaan ingin memiliki kemampuan untuk memprediksi kemungkinan pelanggan yang akan *churn*, sehingga tim promosi dapat memberi penawaran promosi khusus pada tipe pelanggan ini.

Selain itu, perusahaan juga ingin mengetahui faktor-faktor yang membuat pelanggan akan *churn*, sehingga tim promosi dapat membuat strategi yang tepat berupa jenis promosi untuk "membujuk" pelanggan supaya tetap menggunakan layanan ecommerse perusahaan ini.

**Analytics Approach**

a. Identifikasi Pola Churn: Menganalisis data historis menggunakan model klasifikasi machine learning untuk mengidentifikasi pola perilaku pelanggan sebelum mereka churn.

b. Faktor-Faktor Pengaruh: Menentukan variabel-variabel yang paling mempengaruhi keputusan pelanggan untuk beralih, seperti Cashback, pengalaman pengguna, atau lamanya penggunaan, dengan menggunakan model klasifikasi untuk penilaian yang lebih presisi.

c. Model Prediktif: Membangun model klasifikasi untuk memprediksi kemungkinan churn pelanggan di masa depan berdasarkan variabel-variabel yang telah diidentifikasi.

**Metrics Evaluation**

Type 1 error : False Positive  
Konsekuensi: Biaya promosi terbuang percuma kepada pelanggan yang sebenarnya tidak Churn

Type 2 error : False Negative  
Konsekuensi: Pelanggan benar-benar Churn dan tidak diberikan treatment apapun berupa promosi 


Pada kasus ecommerce, jika tim promosi salah membaca customer behavior dan memberikan promosi secara tidak tersegmentasi, akan memicu kecenderungan perilaku pelanggan yang akan terlalu bergantung pada promosi. Tanpa promosi, mereka cenderung tidak bertransaksi karena terlalu sering 'dimanjakan' dengan promosi. Hal ini dapat memberatkan biaya (cost) perusahaan.

Berdasarkan konsekuensi tersebut, tim promosi mengharapkan sebisa mungkin tim data scientist membuat model yang dapat mengurangi biaya promosi dari perusahaan, tetapi tidak membuat angka Churn terlalu tinggi. Dengan demikian, tim promosi ingin sebanyak mungkin prediksi kelas positif yang benar, dengan sesedikit mungkin prediksi false positive. Akhirnya, metric utama yang dapat digunakan adalah roc_auc.

ROC merupakan singkatan dari Receiver Operating Characteristic. Kurva ROC akan melihat trade off antara True Positive  Rate (TPR) dengan False Positive Rate (FPR). Metrik ini digunakan untuk mengevaluasi kinerja model klasifikasi, terutama dalam konteks ini adalah analisis churn. ROC AUC memberikan gambaran tentang sejauh mana model klasifikasi mampu membedakan antara kelas positif (pelanggan yang akan Churn) dan kelas negatif (pelanggan yang akan tetap menggunakan layanan).

## C. Data Cleaning

#### - *Outliers Checking*

Telah disebutkan bahwa tidak tampak adanya nilai *outliers* pada data berdasarkan eksplorasi data menggunakan metode describe() ketika dikomparasi antara nilai mean dan mediannya. Namun, hal tersebut butuh dipastikan. Sebab, nilai *ourliers* yang ekstrim dapat menyebabkan prediksi menjadi bias. Salah satu caranya adalah penggunaan teknik IQR. IQR sendiri bekerja dengan menentukan batas atas dan batas bawah dari suatu distribusi data. Data-data yang terletak di luar batas akan dikategorikan sebagai *outliers*. 

Sebelum memeriksa jumlah outliers melalui IQR, sebaiknya data dieksplorasi terlebih dahulu dengan visualisasi distplot untuk melihat sebaran data, sekaligus juga identifikasi secara visual kemungkinan adanya outliers.

```python
data_numeric = ['Tenure', 'WarehouseToHome', 'NumberOfDeviceRegistered',
       'SatisfactionScore', 'NumberOfAddress', 'Complain', 'DaySinceLastOrder',
       'CashbackAmount']
```

```python
# Data Cenderung Right Skew
plt.figure(figsize=(12, 6))

for i, column in enumerate(data_numeric, start=1):
    plt.subplot(4, 2, i)
    sns.distplot(df[column])

plt.tight_layout()
plt.show()
```

```python
# Dibuktikan melalui uji normalitas
for column in data_numeric:
    statistic, p_value = kstest(df[column], 'norm')

    # Interpretasi p-value
    alpha = 0.05
    if p_value > alpha:
        print(f"Data '{column}' terdistribusi normal (tidak dapat menolak H0)\n")
    else:
        print(f"Data '{column}' tidak terdistribusi normal (menolak H0)")
```

```python
# Menghitung prosentase Outliers
# Menghitung jumlah dan persentase outliers dengan metode IQR

for column in data_numeric:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    print(f'Outliers pada {column} adalah {outliers.count()} atau sebesar {round(outliers.count()/df.shape[0]*100,2)}%')
```

```python
df_outliers = df.copy()

for column in data_numeric:
    Q1 = df_outliers[column].quantile(0.25)
    Q3 = df_outliers[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    
    df_outliers = df_outliers[(df_outliers[column] >= lower_bound) & (df_outliers[column] <= upper_bound)]

df_outliers.reset_index(drop=True, inplace=True)
```

```python
# Before Remove Outliers
print(f'Data Proportion Before Removing Outliers:\n{df["Churn"].value_counts().sort_index()/df.shape[0]*100}\n')
print(f'Data Proportion Before Removing Outliers:\n{df_outliers["Churn"].value_counts().sort_index()/df_outliers.shape[0]*100}\n')
print(f'Data hilang sebesar {round((df.shape[0] - df_outliers.shape[0])/df.shape[0]*100,2)}% akibat removing outliers')
```

Pada tahap ini juga dilakukan uji normalitas untuk menentukan apakah data tersebar secara normal atau tidak. Hasilnya, semua data tidak tersebar secara normal. Dan dikomparasi dengan visualisasi, sebagian besar data tersebar secara right skew, kecuali kolom CashbackAmount.

Proses dilanjutkan dengan metode IQR untuk melihat total data outliers. Outliers terbesar terjadi pada kolom CashbackAmount dengan jumlah 316 data atau sebesar 8.02%, kedua adalah kolom NumberOfDeviceRegistered berjumlah 271 data atau sebesar 6.88%. Salah satu langkah penanganan outliers adalah removing (penghapusan data). Sayangnya, penghapusan data data menyebabkan perubahan komposisi data dan menyebabkan prediksi dari model menjadi bias dari kondisi asli data. Oleh sebab itu, cara lain perlu diperiksa untuk menghindari kondisi bias tersebut.

Pertama adalah membuat simulasi penghapusan data, kemudian liat dampaknya. Penghapusan data akibat outliers menyebabkan hilangnya banyak informasi sebesar 30.47% data, sekaligus merubah proporsi dari target. Oleh karena sebaran data tersebut masih dapat dianggap wajar (data sebenarnya) dan outliers tidak terjadi secara ekstrim, maka outliers akan dipertahankan. Strategi yang dapat digunakan ketika dilakukan modeling nanti adalah dengan penggunaan metode scaler, sehingga datanya dinormalisasikan. ## E. Data Preparation

#### - *Data Engineering*

Modeling Machine Learning tidak dapat membaca data yang bersifat kategori. Konversi data kategori menjadi numerik merupakan hal yang penting dilakukan pada tahap persiapan data. Pada kasus data ecommerce churn ini, terdapat dua data yang bersifat kategori, yaitu 'PreferedOrderCat' dan 'MaritalStatus'. Berdasarkan total data unik, karena MaritalStatus hanya berjumlah tiga, sehingga akan lebih efektif menggunakan metode One Hot Encoding, sedangkan PreferedOrderCat menggunakan Binary Encoding untuk menghindari overfitting karena terlalu banyak feature yang direkayasa.

Sesuai penjelasan pada subbab outlier, metode Scaler Robust akan digunakan agar model dapat beradaptasi dengan kondisi outlier pada data ini. Selain itu, Robust scaler juga merupakan metode scaler yang paling tahan terhadap outliers.

```python
for col in ['PreferedOrderCat', 'MaritalStatus']:
    print(f'Value counts of {col}:')
    unique_data = df[col].value_counts().sort_index()
    data_pct = pd.DataFrame({'total_data': unique_data, 'Pct':round((unique_data/len(df))*100,2)})
    print(data_pct)
    print('\n')
```

```python
scaler_var = ['Tenure', 'WarehouseToHome', 'NumberOfAddress', 'DaySinceLastOrder', 'CashbackAmount']

transformer = ColumnTransformer([
    ('scaler', RobustScaler(), scaler_var),
    ('one hot encoder',OneHotEncoder(drop = 'first'),['MaritalStatus']),
    ('binary encoder',ce.BinaryEncoder(),['PreferedOrderCat'])
], remainder = 'passthrough')
```

#### - *Data Splitting and Transformer*

```python
# Menentukan data features dan data target (Churn)
X = df.drop(columns=['Churn'])
y = df['Churn']
```

```python
# Melakukan splitting data
X_train_val, X_test, y_train_val, y_test = train_test_split(
    X,
    y,
    stratify=y,
    test_size=0.2,
    random_state=42)
```

```python
# Melihat hasil dari feature engineering yang dilakukan dengan metode transformer
features = pd.DataFrame(transformer.fit_transform(X_train_val),
                        columns=transformer.get_feature_names_out())
features
```

```python
# Kolom-kolom hasil feature engineering
features.columns.tolist()
```

#### - *Features Selection*

Pada model machine learning, feature yang terlalu banyak dapat menyebabkan model menjadi overfitting. Namun, sedikit feature juga akan menyebabkan kondisi underfitting karena dinilai terlalu sederhana. Oleh sebab itu, setelah feature diolah dengan transformer, penting untuk menyeleksi fitur-fitur mana saja yang akan berkontribusi terhadap target (churn). Sejumlah feature dengan score teratas akan digunakan untuk diterapkan ke dalam model machine learning yang akan dibuat. Salah satu metode seleksi fitur yang umum digunakan adalah SelectPercentile()

Selain itu sebelum masuk ke proses seleksi fitur, melihat hubungan antar fitur juga penting dilakukan untuk mengetahui fitur-fitur mana yang memiliki hubungan kuat dengan fitur target, dan fitur-fitur lain yang saling berhubungan. Metode korelasi yang umum digunakan adalah dengan metode corr(), dan metode yang digunakan adalah spearman karena data tidak terdistribusi normal atau tidak linear.

```python
plt.figure(figsize = (12,8))
corr = df.corr(method='spearman')
heatmap = sns.heatmap(corr, annot=True)

plt.title('Variable Correlation', fontsize=11)
plt.show()
```

```python
X_train_preprocessed = transformer.fit_transform(X_train_val)
X_test_preprocessed = transformer.transform(X_test)

feature_selection = SelectPercentile(score_func=f_classif, percentile = 55)
feature_selection.fit(X_train_preprocessed, y_train_val)

X_train_preprocessed_selected = feature_selection.transform(X_train_preprocessed)
X_test_preprocessed_selected = feature_selection.transform(X_test_preprocessed)

print('before selection',X_train_preprocessed.shape)
print('after selection',X_train_preprocessed_selected.shape)

mask = feature_selection.get_support()
feature_scores = feature_selection.scores_.round(2)

selected_features = pd.DataFrame({
    'indicator':mask,
    'features':features.columns,
    'score': feature_scores
}).sort_values(by='score', ascending=False)

selected_features[selected_features['indicator']==True]
```

Ketujuh fitur tersebut memiliki score teratas. Sebab score sebelum scaler_CashbackAmount adalah 45, sehingga skor terbawah dibatasi pada nilai 79.

Selanjutnya pada data set, model diseleksi dengan menguji kualitas model pada data baru, dalam hal ini X_test. Metrix ROC AUC digunakan karena ingin sebanyak mungkin prediksi kelas positif yang benar, dengan sesedikit mungkin prediksi false positive sesuai problem statement yang telah disampaikan.

#### - *Model Selection: SK-Fold pada Data Training*

```python
logreg = LogisticRegression(random_state=2024)
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier(random_state=2024)
rf = RandomForestClassifier(random_state=2024)
xgb = XGBClassifier(random_state=2024)
adb = AdaBoostClassifier(random_state=2024)
gbc = GradientBoostingClassifier(random_state=2024)
```

```python
model_var = [logreg, knn, tree, rf, xgb, adb, gbc]
model_name = ['Logistic Regression',
              'KNN',
              'Decision Tree',
              'Random Forest',
              'XGBoost',
              'Adaboost',
              'Gradient Boosting']

score=[]
mean=[]
std=[]

for i in model_var:
    skfold = StratifiedKFold(n_splits = 5, random_state = 42, shuffle=True)
    estimator = Pipeline([
        ('preprocess', transformer),
        ('feature_selection', feature_selection),
        ('clf', i)
    ])
    model_cv=cross_val_score(estimator, X_train_val, y_train_val, cv=skfold, scoring='roc_auc')
    score.append(model_cv)
    mean.append(model_cv.mean())
    std.append(model_cv.std())
    
pd.DataFrame({'model':model_name,
              'mean roc_auc':mean,
              'sdev':std}).set_index('model').sort_values(by='mean roc_auc',ascending=False)
```

```python
for model, s in zip(model_name, score):
    print(f"{model}:")
    print(f"{s}\n")
```

#### - *Model Selection: Test Data*

```python
model_var = [logreg, knn, tree, rf, xgb, adb, gbc]

model_name = ['Logistic Regression',
              'KNN',
              'Decision Tree',
              'Random Forest',
              'XGBoost',
              'Adaboost',
              'Gradient Boosting']

score_roc_auc = []

def y_pred_func(model):
    estimator = Pipeline([
        ('preprocess', transformer),
        ('feature_selection', feature_selection),
        ('clf', model)])
    X_train_val, X_test
    estimator.fit(X_train_val, y_train_val)
    return(estimator,estimator.predict(X_test),X_test)

for model, name in zip(model_var, model_name):
    estimator, y_pred, X_test = y_pred_func(model)
    y_predict_proba = estimator.predict_proba(X_test)[:,1]
    score_roc_auc.append(roc_auc_score(y_test, y_predict_proba))
    print(name,'\n', classification_report(y_test, y_pred))
    
pd.DataFrame({'model':model_name,
             'roc_auc score':score_roc_auc}).set_index('model').sort_values(by='roc_auc score',ascending=False)
```

2. Gradient Boosting

Model terbaik berdasarkan test set:
1. Random Forest

3. XGBoost

Terlihat bahwa pada Cross validasi, XGBoost merupakan model terbaik. Namun, posisinya turun menjadi diurutan ketiga pada data test. Sedangkan, Gradient Boosting stabil di posisi kedua. Jika dibandingkan ketiga model teratas, Gradient Boosting juga memiliki nilai standard deviasi yang kecil yakni, 0.021670. Berdasarkan pertimbangan tersebut, model yang akan dipilih adalah Gradient Boosting.

Dikutip dari www.trivusi.web.id, Gradient Boosting adalah sebuah teknik ensemble yang menggabungkan beberapa model yang lemah (weak model) menjadi sebuah model yang kuat. Model-model lemah ini sering disebut dengan weak learners, dan dapat berupa model regresi atau klasifikasi sederhana seperti Decision Tree. Algoritma ini menggunakan pendekatan iteratif, di mana setiap iterasi bertujuan untuk meningkatkan model sebelumnya dengan menambahkan model baru. Proses ini dilakukan secara berulang-ulang hingga model yang dihasilkan memenuhi kriteria tertentu, seperti nilai loss function yang cukup kecil. 

Proses iteratif dalam algoritma Gradient Boosting terdiri dari beberapa tahap, yaitu:
1. Inisialisasi model
2. Membuat weak model
3. Menghitung residual error
4. Menyusun kembali data training
5. Membuat model baru
6. Menggabungkan model
7. Iterasi berulang

Gradient Boosting ini sangat cocok diterapkan pada analisis Churn. Gradient Boosting dapat menangani hubungan non-linier antara fitur dan variabel target. Hal ini dapat berguna saat mencari pola kompleks yang mungkin terkait dengan perilaku pelanggan yang berpotensi berhenti. Selanjutnya, jika data pelanggan terstruktur dengan baik, Gradient Boosting dapat memberikan hasil yang akurat. Ini termasuk variabel-variabel seperti frekuensi penggunaan, lamanya penggunaan, dan fitur-fitur lainnya yang dapat diambil dari data pelanggan. Terakhir, Gradient Boosting dapat menangani baik variabel target kategorikal maupun kontinu. Ini berguna jika variabel target churn bersifat biner (misalnya, churn atau tidak churn). Selain itu, outliers dan missing values juga telah ditangani dengan baik, sehingga dapat lebih mendukung kinerja model Gradient Boosting ini. Juga, score metrik ROC_AUC pada model evaluasi Cross Validation dan test set juga stabil sehingga meminimalisir kemungkinan terjadinya overfitting.

#### - *Hyperparameter Tuning: Mengoptimalkan Evaluation Metrik*

Setelah model diperoleh, langkah selanjutnya adalah mengoptimalkan performa model yang telah dipilih. Salah satu caranya adalah dengan melakukan hyperparameter tuning sebagai berikut.

```python
gbc = GradientBoostingClassifier(random_state=2024)

estimator = Pipeline([
    ('preprocess', transformer),
    ('feature_selection', feature_selection),
    ('clf', gbc)])

hyperparam_space = {
    'clf__n_estimators': [50, 100, 200],
    'clf__learning_rate': [0.01, 0.1, 0.2],
    'clf__max_depth': [3, 5, 7]
}

grid_search = GridSearchCV(
    estimator, # model to tune
    param_grid = hyperparam_space, # hyperparameter space
    cv = skfold, # evaluation method
    scoring = 'roc_auc', # metrics
    n_jobs = -1 # use all cores
)

```

```python
grid_search.fit(X_train_val, y_train_val)
```

```python
print('best score',grid_search.best_score_)
print('best param',grid_search.best_params_)
```

```python
gs_result = pd.DataFrame(grid_search.cv_results_)
gs_result[gs_result['rank_test_score']<5]
```

Berdasarkan hyperparameter tuning yang telah dilakukan, didapatkan parameter terbaik sebagai berikut

`best param {'clf__learning_rate': 0.1, 'clf__max_depth': 5, 'clf__n_estimators': 200}`

dengan skor ROC_AUC terbaik sebesar 0.92 atau 92%. Selanjutnya adalah membuat perbandingan performa antara sebelum dilakukan tuning dan setelah dilakukan tuning.

Hal tersebut didasarkan pada kondisi dimana tim promosi mengharapkan sebisa mungkin model yang dibuat dapat mengurangi biaya promosi dari perusahaan, tetapi tidak membuat angka Churn terlalu tinggi. ROC merupakan singkatan dari Receiver Operating Characteristic. Kurva ROC akan melihat trade off antara True Positive Rate (TPR) dengan False Positive Rate (FPR). Metrik ini digunakan untuk mengevaluasi kinerja model klasifikasi, terutama dalam konteks ini adalah analisis churn. ROC AUC memberikan gambaran tentang sejauh mana model klasifikasi mampu membedakan antara kelas positif (pelanggan yang akan Churn) dan kelas negatif (pelanggan yang akan tetap menggunakan layanan).

Metrik lain yang penting diperhatikan adalah recall. Recall merupakan rasio prediksi benar positif dibandingkan dengan keseluruhan data yang benar positif. Recall menjawab pertanyaan â€œBerapa persen customer yang diprediksi Churn dibandingkan keseluruhan customer yang sebenarnya churnâ€.
